```python
import math
def func(x):
    return (1/(1+x*x))

def Simpsons_Rule(x0, xn, n):
    h = (xn-x0)/n
    total = func(x0) + func(xn)
    xi = 0.0
    for i in range (1,n):
        xi = x0 + h
        if (i%2 == 0):
            total += 2*func(xi)
        else:
            total += 4*func(xi)
        x0 = xi

    return (h/3) * total

x0  = float(input("Enter the lower limit: "))
xn  = float(input("Enter the upper limit: "))
n  = int(input("Enter the number of interval: "))

approx_area = Simpsons_Rule(x0, xn, n)
print("=> By using Simpsons method: ")
print(f"The obtained approximate area is: {approx_area:.4f}")

exact_sol = math.atan(xn) - math.atan(x0)
print(f"The actual area is: {exact_sol:.4f}")
error = abs(exact_sol - approx_area)
print(f"The error is: {error:.4f}")
```

```python
import math
import pandas as pd

def func(x):
    return 1 / (1 + x*x)

def Simpsons_Rule(x_values, y_values):
    n = len(x_values) - 1
    if n % 2 != 0:
        raise ValueError("Number of subintervals (n) must be even for Simpson's 1/3 Rule.")
    
    h = x_values[1] - x_values[0]
    total = y_values[0] + y_values[-1]

    for i in range(1, n):
        if i % 2 == 0:
            total += 2 * y_values[i]
        else:
            total += 4 * y_values[i]

    return (h / 3) * total

x_values = [0, 0.5, 1, 1.5, 2, 2.5, 3]
y_values = [func(x) for x in x_values]

data = {'x': x_values, 'f(x)': y_values}
table = pd.DataFrame(data)
print("Table of Values:")
print(table.to_string(index=False))

approx_area = Simpsons_Rule(x_values, y_values)

x0, xn = x_values[0], x_values[-1]
exact_sol = math.atan(xn) - math.atan(x0)
error = abs(exact_sol - approx_area)

print("\n=> Using Simpson’s 1/3 Rule:")
print(f"Approximate area = {approx_area:.6f}")
print(f"Exact area       = {exact_sol:.6f}")
print(f"Error            = {error:.6f}")
```
